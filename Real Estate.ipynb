```python
def calc_payment(P, i, N):
    i = i/12
    n = i*(1+i)**N
    d = (1+i)**N-1
    return P*(n/d)
calc_payment(400000, .035, 30*12)

def calc_interest(P, i):
    i = i/12
    return 1+i*P
```

```python
def cumulative_value(df, rate=.05):
    periods = len(df)    
    n_pers = (periods - pd.Series(range(len(df))) -1).values
    return sum(df*exp(rate/12.*n_pers))

#cumulative_value(df.savings)

def make_data(buying_price, downpayment=.2, interest_rate=3.5, 
                  years=30, home_appreciation=5, property_tax=1.5,
                  closing_costs=4, monthly_maintenance=100, insurance=.45,
                  assess_frequency=5, inflation=2.0,
                  investment_return=5.,
                  rent=3000):
    principal = buying_price*(1-downpayment)
    p = principal
    
    payment = calc_payment(principal, interest_rate/100, years*12)
    
    data = []
    
    for i in range(0, years*12):
        interest = calc_interest(p, interest_rate/100)
        p -= payment
        p += interest
        data.append([i, interest, payment, p])
    
    df = pd.DataFrame(data, columns=['period', 'interest', 'payment', 'balance'])
    df['principal_reduction'] = df.payment - df.interest
    df['maintenance'] = monthly_maintenance *exp(inflation/100./12*df.period)
    df['rent'] = rent
    
    df['value'] = buying_price * exp(home_appreciation/100./12*df.period)
    df['insurance'] = df.value * (insurance/100/12)
    #In SF property tax increases are capped at 2% per year
    #Assessment may not occur every year.
    df['assess'] = df.period % (12*assess_frequency) == 0
    df['property_tax'] = np.NaN
    df['property_tax'] = df.property_tax.where(~df.assess, df.value*property_tax/100/12)
    df['property_tax'] = df.property_tax.fillna(method='ffill')
    df['property_tax'] = df.property_tax.where((df.property_tax.pct_change(12) <= .02) | df.property_tax.pct_change(12).isnull(),
                                                df.property_tax.shift(12)*1.02)
    df['paid_to_date'] = buying_price*(downpayment) + df.payment.cumsum() + df.property_tax.cumsum() + \
                          closing_costs/100*buying_price + df.maintenance.cumsum()
    df['equity'] = df.value - df.balance
    df['ROI'] = (df.equity - df.paid_to_date) / df.paid_to_date * 100
    df['savings'] = (closing_costs/100*buying_price+buying_price*downpayment)*exp(investment_return/100/12*df.period)
    df['rent_savings'] = df.payment + df.maintenance + df.property_tax + df.insurance - df.rent
    df['time_rent_savings'] = pd.stats.moments.expanding_apply(df.rent_savings, cumulative_value)
    df['total_savings'] = df.savings + df.time_rent_savings
    
    return df
```

```python
def make_interact(buying_price, downpayment=.2, interest_rate=3.5, 
                  years=30, home_appreciation=5, property_tax=1.5, closing_costs=4.,
                  monthly_maintenance=(100,1000,100), investment_return=5., rent=35000):
    
    df = make_data(buying_price, downpayment, interest_rate, 
                  years, home_appreciation=home_appreciation, 
                  property_tax=property_tax, closing_costs=closing_costs,
                  investment_return=investment_return, rent=rent)    
    
    plt.figure(); df[['equity', 'paid_to_date', 'total_savings']].plot(figsize=(24,8))
    plt.figure(); df.ROI.plot(figsize=(24,8))
    draw_bars(df[df.period%4 == 0]).show()

i = interact(make_interact,
         buying_price=(400000,1200000),
         downpayment = (.1, .5, .01),
         interest_rate=(0.,10.,0.01),
         years=(25,35,5),
         home_appreciation=(-10.,10.,.1),
         property_tax=(0, 10, .1),
         closing_costs=(0,10,.1),
         rent=(0,7000,50)
         )
#df = make_data(800000, .035)
```
